Reindexing columns

```python
df = df.reindex(columns=['Happiness Score', 'Region', 'Standard Error',
       'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual'])
df.head()
```

```python
##Assigning x and y values to pass to the model
x=df.iloc[:,1:]
x
```

```python
##Assigning the happiness score to y
y=df.iloc[:,0]
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.linear_model import Lasso, Ridge
from sklearn.svm import SVR
```

```python
#splitting for training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=43)
```

```python
def model(mod):
    mod.fit(x_test,y_test)
    ##mod.coef_
    pred=mod.predict(x_test)
    mod.score(x_train,y_train)
    print('\n\n')
    ##print('predicted', pred)
    ##print('actual:',y_test)
    rscr=r2_score(y_test,pred)
    print('r2score:',r2_score(y_test,pred))
    print('Mean squared error',mean_squared_error(y_test,pred))
    print('mean absolute error:',mean_absolute_error(y_test,pred))
    
```

```python
#function for cross-validation
from sklearn.model_selection import cross_val_score
def crossval(mod):
    scores = cross_val_score(mod, x_train, y_train, scoring='r2', cv=5)
    print(scores)
    mean=scores.mean()
    print("Mean:",mean)
    
```

### Linear Regression

```python
lr=LinearRegression()
model(lr)
crossval(lr)
```

### Lasso

```python

lasso=Lasso(alpha=0.01)
model(lasso)
crossval(lasso)
```

### SVR

```python
k=['linear','poly','rbf']
for i in k:
    print('SVR kernel:',i)
    print('\n')
    svr=SVR(kernel=i)
    model(svr)
    crossval(svr)
```
